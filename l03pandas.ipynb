# Задание 2

```python
def floatDistance(s):
    if "/" in s:
        ss = s.split("/")
        return float(ss[1])
    elif s!="-":
        return float(s)
    else:
        return float(-1)

df['Расст до Якт наземным'] = df['Расст до Якт наземным'].apply(floatDistance)
df['Расст до Якт водным'] = df['Расст до Якт водным'].apply(floatDistance)
df['Расст до Якт воздушным'] = df['Расст до Якт воздушным'].apply(floatDistance)

df['ratio'] = df['Расст до Якт наземным']/df['Расст до Якт воздушным']
df
```

```python
df['Улус'][df['ratio'].idxmax()]
```
